```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import math
from ripser import ripser
from persim import plot_diagrams


import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault) # otherwise, get latex error when visualizing plots
```

# Create a more complicated stimulus space

```python
def remove_ellipse(PC, x0, y0, a, b, r):
    # given PC, remove points in the ellipse ((x-x0)/a)^2 + ((y-y0)/b)^2 <= r^2
    n = PC.shape[0]

    # index of points to remove
    remove_idx = []
    for i in range(n):
        x = PC[i,0]
        y = PC[i,1]

        if ((x - x0)/a)**2 + ((y-y0)/b)**2 <= r**2:
            remove_idx.append(i)

    return np.delete(PC, remove_idx, 0)
```

```python
n_PC1 = 700
PC1 = np.random.rand(n_PC1,2)

# punch one ellipse centered at (0.2, 0.4).  
PC1 = remove_ellipse(PC1, 0.2, 0.3, 0.9, 1.4, 0.15)

# punch one large ellipse centered at (0.8, 0.8) that covers 3 ellipses in PC1
PC1 = remove_ellipse(PC1, 0.75, 0.75, 1, 1.2, 0.18)

# punch one large ellipse that doesn't exist in PC1
PC1 = remove_ellipse(PC1, 0.2, 0.8, 1, 0.9, 0.14)

# punch two holes that have ambiguous matches
PC1 = remove_ellipse(PC1, 0.7, 0.28, 0.9, 0.4, 0.18)
PC1 = remove_ellipse(PC1, 0.7, 0.1, 0.9, 0.4, 0.18)

print("number of points in PC1: ", PC1.shape[0])
```

```python
n_PC2 = 1000
PC2 = np.random.rand(n_PC2,2)

# punch one large ellipse centered at (0.2, 0.4)
PC2 = remove_ellipse(PC2, 0.2, 0.3, 0.9, 1.4, 0.15)

# punch three discs near each other
PC2 = remove_ellipse(PC2, 0.6, 0.8, 0.4, 0.4, 0.18)

# punch three discs near each other
PC2 = remove_ellipse(PC2, 0.8, 0.6, 0.7, 0.4, 0.18)

# punch three discs near each other
PC2 = remove_ellipse(PC2, 0.85, 0.85, 0.3, 0.7, 0.18)

# punch the "ambigous" hole
PC2 = remove_ellipse(PC2, 0.7, 0.18, 0.9, 0.4, 0.2)

print("number of points in PC2: ", PC2.shape[0])
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (10,5))
ax[0].scatter(PC1[:,0], PC1[:,1], color = "#00b4bc", s = 3)
ax[0].set_title("point cloud 1")
ax[1].scatter(PC2[:,0], PC2[:,1], color = "#ee9c33", s = 3)
ax[1].set_title("point cloud 2")
#plt.savefig("pointclouds_2d.pdf")
plt.show()
```

```python
persistence1 = ripser(PC1,
                    maxdim=1,
                    coeff=47,
                    do_cocycles=True,
                    )
```

```python
plt.figure()
plot_diagrams(persistence1["dgms"])
plt.show()
```

```python
persistence2 = ripser(PC2,
                    maxdim=1,
                    coeff=47,
                    do_cocycles=True,
                    )
```

```python
plt.figure()
plot_diagrams(persistence2["dgms"])
plt.show()
```

```python
# save PC1, PC2
np.savetxt("data/complicated_stimulus_space/pointcloud1.csv", PC1, delimiter = ",")
#np.savetxt("data/complicated_stimulus_space/pointcloud2.csv", PC2, delimiter = ",")
```

# Make both point clouds to live on a curvy surface

```python
PC1_z = (np.cos(PC1[:,0] * 2 * math.pi ) * np.cos(PC1[:,1] * 2 * math.pi) * 0.05) + 0.2
PC1_z = PC1_z.reshape(-1, 1)
PC1_3d = np.concatenate((PC1, PC1_z), axis = 1)

PC2_z = np.cos(PC2[:,0] * math.pi) * np.sin(PC2[:,1] * 2 * math.pi) * 0.05
PC2_z = PC2_z.reshape(-1,1)
PC2_3d = np.concatenate((PC2, PC2_z), axis = 1)
```

```python
n_PC1 = PC1.shape[0]
n_PC2 = PC2.shape[0]


fig = go.Figure(data = [go.Scatter3d(x = PC1_3d[:,0], y = PC1_3d[:,1], z = PC1_3d[:,2],  
                                    mode='markers',
                                    marker=dict( size=2,color = "#00b4bc")
                                    )])

fig.add_trace(go.Scatter3d(x = PC2_3d[:,0], 
                            y = PC2_3d[:,1], 
                            z = PC2_3d[:,2],
                            mode='markers',
                            marker = dict(size = 2, color = "#ee9c33") )        
            )


fig.show()
```

## Check PH

```python
persistence1 = ripser(PC1_3d,
                    maxdim=1,
                    coeff=47,
                    do_cocycles=True,
                    )
```

```python
plt.figure()
plot_diagrams(persistence1["dgms"])
plt.show()
```

```python
persistence2 = ripser(PC2_3d,
                    maxdim=1,
                    coeff=47,
                    do_cocycles=True,
                    )
```

```python
plt.figure()
plot_diagrams(persistence2["dgms"])
plt.show()
```
